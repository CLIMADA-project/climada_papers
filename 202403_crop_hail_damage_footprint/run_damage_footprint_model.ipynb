### A crop hail damage footprint model at your finger tips

This Notebook models crop hail damage footprints from radar using CLIMADA. It allows to set user-specific MESHS thresholds and spatial resolutions as well as the minimum number of fields at each grid cell for it to count as exposure ($n_{thresh}$).

```python
#own
import sys
from pathlib import Path
#add path to path of current skript, where also utility.py lies
sys.path.append(Path.cwd().absolute())
#own utility functions
from utility import aggregate_hazard, exposure_from_gpd,\
                    read_gridded_damages_as_impact, hazard_from_radar,\
                    compute_verification_stats, en_names

from calibration import empirical_calibration_per_exposure, DROP_DATES_DICT, drop_events

# climada
from climada.util.save import load
from climada.entity import ImpactFunc, ImpactFuncSet
from climada.engine import ImpactCalc, Impact

# others
import numpy as np
import pandas as pd

from climada.util.api_client import Client


```

First, we specify MESHS threshold, resolution and $n_{thresh}$ as well as the croptype that we would like to work with

```python
MESHS_threshold=30 #in mm
resolution=8 #in km
croptype='wheat'
n_thresh=20 #in number of fields
```

Then, we get the hail hazard data at 1km resolution from the Climada Data-API and coarsen the resolution thereafter. For the aggregation we use the maximum value within a grid cell.

```python
#get hail hazard data at 1km
client=Client()
hazard_1km=client.get_hazard('hail', properties={'variable': 'MESHS'})
```

```python
hazard_1km.plot_intensity(event='ev_2021-06-21')
```

```python
#spatially aggregate to target grid
hazard, _=aggregate_hazard(hazard_1km, original_grid_epsg = 2056, 
                                  extent_new = None, 
                                  cell_size_new = resolution*1000,
                                  projection_new_epsg = 4326,
                                  aggfunc = 'max')
```

So lets now plot the aggregated hazard data for an individual hail date

```python
hazard.plot_intensity(event='ev_2021-06-21')
```

Next, we read the gridded exposure data from the Climada Data-API. For that, we need to additionally specify the croptype. We can also set a minimum number of exposed fields $n_{thresh}$ for a grid point to count as exposure. The exposure value is here a binary field (yes/no) but can be changed to the number of fields (`n_fields`) or to the total area in ha (`area_ha`), which are available as separate columns in the `exposure.gdf`. For the simple model applied here the binary field is used. The value unit is set to `"yes (1) no (0)"`

```python
exposure = client.get_exposures('crops', properties={'res_km': f'{resolution}', 'crop': croptype})

exposure.gdf['value'][exposure.gdf['n_fields']<n_thresh]=0
exposure.value_unit='yes (1) no (0)'
```

Now, lets plot the exposure as well

```python
exposure.plot_hexbin(gridsize=100)
```

We can also investigate some other aspects of our exposure data from information passed on in the gdf (such as the number of fields or the area)

Next we define the impact function, which is a step function defined by threshold s

```python
# We initialise an ImpactFunc for hail damage to crops.
def step_function(x,threshold):
    y=np.where(x<threshold,0,1)
    return y

haz_type = "HL"
id = 1
name = "HL agriculture damage"
# provide unit of the hazard intensity
intensity_unit = "mm"
# provide values for the hazard intensity range
intensity = np.arange(20, 101, 1)
paa=step_function(intensity,MESHS_threshold)
mdd = np.ones(len(intensity))
imp_fun = ImpactFunc(
    id=id,
    name=name,
    intensity_unit=intensity_unit,
    haz_type=haz_type,
    intensity=intensity,
    mdd=mdd,
    paa=paa,
)

# check if the all the attributes are set correctly
imp_fun.check()
impf_set = ImpactFuncSet([imp_fun])
impf_set.check()
imp_fun.plot()
```

Now, we are ready to compute the impact (i.e. the damage footprints). 

```python
impact = ImpactCalc(exposure, impf_set, hazard).impact(save_mat=True)
```

Now, lets plot the impact for a single event. Aggregated in a hexbin plot the impact is the number of grid points with damage expected. Note that the selection of the gridsize argument is important here. To plot the individual grid cells as hexbins with a spatial resolution of 8km, `gridsize=100` is a good choice.

```python
impact_now=impact.select(event_names=['ev_2021-06-21'])
impact_now.plot_hexbin_impact_exposure(event_id=impact_now.event_id[0], gridsize=100)
```

Now, we can read the damage data and compare it visually to the modelled impact (damage footprints). As the impact form the CLIMADA-API indicates the number of damaged fields we create a new impact object called `damages_binary` with a binary impact matrix (1 if number of damaged fields>0 and 0 otherwise).

```python
ds = client.get_dataset_info(data_type='hail_damage_crops', properties={
    'res_km': f'{resolution}',
    'crop': croptype,
})
_, [impact_file] = client.download_dataset(ds)
 
damages = Impact.from_hdf5(impact_file)
from copy import deepcopy
damages_binary=deepcopy(damages)

#replace number of fields with 1 where number of fields is larger than 1
indices=damages_binary.imp_mat.nonzero()
imp_mat_binary=damages_binary.imp_mat.copy()
imp_mat_binary[indices]=1
damages_binary.imp_mat=imp_mat_binary
damages_binary.unit='yes (1) no (0)'

#select and plot damages
damages_now=damages_binary.select(event_names=['ev_2021-06-21'])
damages_now.plot_hexbin_impact_exposure(event_id=damages_now.event_id[0], gridsize=100)
```

Now, how do verification scores look like for this data? For this we can use the `empirical_calibration_per_exposure` function which matches hazard, exposure and damages at the hazard centroids. This can be done for either one single event (use `damages_now`) or all events (use `damages`). The results will be valid at the hazard grid (e.g. 8km). Exposure and damages can be passed also at higher resolution grids, this does not change the result.

If we use all events, it is important to remove the events that occurred when the particular crop has already been harvested. These dates per crop are stored in the `DROP_DATES_DICT` dictionary. 

```python
#drop damages at dates where crop was already harvested
events=list(damages.event_name)
for ev in DROP_DATES_DICT[croptype]:
    if ev in events:
        events.remove(ev)
damages=damages.select(event_names=events)

_, _, _, _, values_at_centroid_all,_ =\
                    empirical_calibration_per_exposure(hazard_object = hazard,
                                                       exposure_object = exposure,
                                                       damages = damages,
                                                       exposure_type = 'agriculture',
                                                       variable = 'MESHS')
```

Then from `values_at_centroid_all`, we can compute verification statistics, based on the indicated n_thresh. The result is a pandas.DataFrame with MESHS values as indices and the scores for this MESHS value as columns. Note that to compute the scores for a MESHS value of e.g. 25 all grid cells with MESHS values larger than or equal to 25 are used.

```python
df, npred = compute_verification_stats(values_at_centroid_all,
                                               variable = 'MESHS',
                                               exposure_thresh=n_thresh)
print(df)
     
```

Finally, we can print the calibration results for the MESHS threshold selected at the beginning of this Notebook.

```python
th=MESHS_threshold
while th not in df.index:
    th+=1

print(f'crop: {croptype}')
print(f'resolution: {resolution}km')
print(f'MESHS threshold: {MESHS_threshold}')
print(f'n_thresh: {n_thresh}')
print(f'POD: {df.POD[th]}, FAR: {df.FAR[th]}, HSS: {df.HSS[th]}, BIAS: {df.B[th]}')
```

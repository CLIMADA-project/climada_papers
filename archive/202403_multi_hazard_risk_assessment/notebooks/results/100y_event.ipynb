```python
import pandas as pd
from pathlib import Path
import itertools

BASE_DIR = Path("/nfs/n2o/wcr/szelie/multi_risk_data")

country = 'global'
warming_levels = ['1']
exposures = ['pop', 'assets']
hazards = ['TC', 'RF']
combinations = list(itertools.combinations(hazards, 2))

# Assuming BASE_DIR and the subsequent directory structure exists and the files are accessible

# Function to open a CSV, modify it, and save it back
def modify_and_save_csv(file_path, haz_type):
    # Open the CSV file into a pandas DataFrame
    df = pd.read_csv(file_path)
    
    # Add a new column 'haz_type' with all NaN values
    df['haz_type'] = pd.NA
    
    # Set the first value of 'haz_type' to the specified hazard type
    if not df.empty:
        df.at[0, 'haz_type'] = haz_type

    # Save the modified DataFrame back to CSV
    df.to_csv(file_path, index=False)

# Loop through the specified conditions and modify the CSV files
for warming in warming_levels:
    for exposure in exposures:
        # Handle compound hazards
        for combi in combinations:
            file_name = f"{exposure}_100y_event_compound_ordered_{'_'.join(combi)}_150arcsec_{warming}_global.csv"
            file_path = BASE_DIR / '100y_event' / file_name
            modify_and_save_csv(file_path, '_'.join(combi))
        
        # Handle single hazards
        for hazard in hazards:
            file_name = f"{hazard}_{exposure}_100y_event_yearsets__150arcsec_{warming}_global.csv"
            file_path = BASE_DIR / '100y_event' / file_name
            modify_and_save_csv(file_path, hazard)
```

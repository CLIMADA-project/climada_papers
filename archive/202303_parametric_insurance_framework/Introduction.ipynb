# Fig. 1: Schematic payout functions
Schematic illustration of a step and linear payout function. Payout functions represent the relationship between a parametric index and a payout. A step payout function defines different parametric index thresholds (thrs) which trigger payout thrs. Linear payout functions assume a linear increase in payout between two index thr. The minimum payout amounts to 0% and the maximum to 100% of the maximum payout

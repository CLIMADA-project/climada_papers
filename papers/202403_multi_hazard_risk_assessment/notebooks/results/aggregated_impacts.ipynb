## We start by loading in the impact objects

```python
import itertools
from climada.engine import Impact
from pathlib import Path

from config import BASE_DIR
BASE_DIR = Path("/nfs/n2o/wcr/szelie/multi_risk_data")
impact_dir = BASE_DIR / 'impact_aggr/global/csv'


country = 'global'
warming_levels = ['1', '2']
exposures = ['assets']
hazards = ['TC', 'RF']
base_dir = Path('/nfs/n2o/wcr/szelie/multi_risk_data/impact_aggr/global/csv')

combinations = list(itertools.combinations(hazards, 2))

def build_impact_path(exposure, combi, suffix, warming):
    filename = f"{exposure}_combined_impact{suffix}_{combi[0]}_{combi[1]}_150arcsec_{warming}_{country}.csv"
    return impact_dir / exposure/ filename

def build_impact_path_not_ordered(exposure, combi, suffix, warming):
    filename = f"{exposure}_combined_impact{suffix}_{'_'.join(combi)}_150arcsec_{warming}_{country}.csv"
    print(filename)
    return impact_dir / exposure / filename

def build_impact_path_caped(exposure, combi, suffix, warming):
    filename = f"{exposure}_combined_impact{suffix}_{combi[0]}_{combi[1]}_150arcsec_{warming}_{country}_caped.csv"
    return impact_dir / exposure/ filename

def buid_impact_path_not_ordered_caped(exposure, combi, suffix, warming):
    filename = f"{exposure}_combined_impact{suffix}_{'_'.join(combi)}_150arcsec_{warming}_{country}_caped.csv"
    return impact_dir / exposure / filename



impact_combined_ordered_not_caped = {
    warming: {
        exposure: {
            combi: Impact.from_csv(
                build_impact_path(exposure, combi, '_ordered', warming)
            )
            for combi in combinations
        }
        for exposure in exposures
    }
    for warming in warming_levels
}

impact_combined_not_ordered_not_caped = {
    warming: {
        exposure: {
            combi: Impact.from_csv(
                build_impact_path_not_ordered(exposure, combi, '', warming)
            )
            for combi in combinations
        }
        for exposure in exposures
    }
    for warming in warming_levels
}


impact_combined_ordered_caped = {
    warming: {
        exposure: {
            combi: Impact.from_csv(
                build_impact_path_caped(exposure, combi, '_ordered', warming)
            )
            for combi in combinations
        }
        for exposure in exposures
    }
    for warming in warming_levels
}

impact_combined_not_ordered_caped = {
    warming: {
        exposure: {
            combi: Impact.from_csv(
                buid_impact_path_not_ordered_caped(exposure, combi, '', warming)
            )
            for combi in combinations
        }
        for exposure in exposures
    }
    for warming in warming_levels
}
```

## The yearsets (or single hazard impact objects)

```python
# Helper function to construct file path
from config import BASE_DIR
BASE_DIR = Path("/nfs/n2o/wcr/szelie/multi_risk_data")
impact_dir = BASE_DIR / 'impact_yearsets/global/csv'

country = 'global'
warming_levels = ['1', '2']
exposures = ['assets']
hazards = ['TC', 'RF']
from climada.engine import Impact

def get_file_path(exposure, hazard, warming, country, capped):
    cap_suffix = "_caped" if capped else ""
    filename = f"{exposure}/{hazard}_{exposure}_impacts_yearsets_150arcsec_{warming}_{country}{cap_suffix}.csv"
    return impact_dir / filename

# Function to load impacts into a nested dictionary structure
def load_impacts(capped):
    return {
        warming: {
            exposure: {
                hazard: Impact.from_csv(get_file_path(exposure, hazard, warming, country, capped))
                for hazard in hazards
            } for exposure in exposures
        } for warming in warming_levels
    }

# Load both capped and not capped impacts
impacts_yearsets_caped = load_impacts(capped=True)
impacts_yearsets_not_caped = load_impacts(capped=False)
```

## We save the difference for the 1/100 year and the average annual impact (AAI) rather the yearly impact is caped at the exposure value or not and rather we consider common drivers or not

```python
rel_diff_dict_100y_not_caped_not_ordered = {
    exp: {
        warming: {
            'imp': (y100_event_dict["combined_not_ordered_not_caped"][exp][warming] - y100_event_dict["combined_ordered_caped"][exp][warming]) / y100_event_dict["combined_ordered_caped"][exp][warming]
        } for warming in y100_event_dict["combined_ordered_caped"][exp]
    } for exp in y100_event_dict["combined_ordered_caped"]
}

rel_diff_dict_100y_not_ordered = {
    exp: {
        warming: {
            'imp': (y100_event_dict["combined_not_ordered_caped"][exp][warming] - y100_event_dict["combined_ordered_caped"][exp][warming]) / y100_event_dict["combined_ordered_caped"][exp][warming]
        } for warming in y100_event_dict["combined_ordered_caped"][exp]
    } for exp in y100_event_dict["combined_ordered_caped"]
}

rel_diff_dict_100y_not_caped = {
    exp: {
        warming: {
            'imp': (y100_event_dict["combined_ordered_not_caped"][exp][warming] - y100_event_dict["combined_ordered_caped"][exp][warming]) / y100_event_dict["combined_ordered_caped"][exp][warming]
        } for warming in y100_event_dict["combined_ordered_caped"][exp]
    } for exp in y100_event_dict["combined_ordered_caped"]
}

rel_diff_dict_aai_not_caped_not_ordered = {
    exp: {
        warming: {
            'imp': (aai_dict["combined_not_ordered_not_caped"][exp][warming] - aai_dict["combined_ordered_caped"][exp][warming]) / aai_dict["combined_ordered_caped"][exp][warming]
        } for warming in aai_dict["combined_ordered_caped"][exp]
    } for exp in aai_dict["combined_ordered_caped"]
}

rel_diff_dict_aai_not_ordered = {
    exp: {
        warming: {
            'imp': (aai_dict["combined_not_ordered_caped"][exp][warming] - aai_dict["combined_ordered_caped"][exp][warming]) / aai_dict["combined_ordered_caped"][exp][warming]
        } for warming in aai_dict["combined_ordered_caped"][exp]
    } for exp in aai_dict["combined_ordered_caped"]
}

rel_diff_dict_aai_not_caped = {
    exp: {
        warming: {
            'imp': (aai_dict["combined_ordered_not_caped"][exp][warming] - aai_dict["combined_ordered_caped"][exp][warming]) / aai_dict["combined_ordered_caped"][exp][warming]
        } for warming in aai_dict["combined_ordered_caped"][exp]
    } for exp in aai_dict["combined_ordered_caped"]
}
```

```python
def flatten_dict(nested_dict):
    res = {}
    if isinstance(nested_dict, dict):
        for k in nested_dict:
            flattened_dict = flatten_dict(nested_dict[k])
            for key, val in flattened_dict.items():
                key = list(key)
                key.insert(0, k)
                res[tuple(key)] = val
    else:
        res[()] = nested_dict
    return res


def nested_dict_to_df(values_dict):
    flat_dict = flatten_dict(values_dict)
    df = pd.DataFrame.from_dict(flat_dict, orient="index")
    df.index = pd.MultiIndex.from_tuples(df.index)
    df = df.unstack(level=-1)
    df.columns = df.columns.map("{0[1]}".format)
    return df
```

```python
import pandas as pd
df = nested_dict_to_df(rel_diff_dict_100y_not_caped_not_ordered).reset_index()
df = df.rename(columns={'level_0':'Exposure','level_1':'Warming', "imp":"Full recovery, no common drivers"})
df["Full recovery, no common drivers"] = df["Full recovery, no common drivers"]*100

df2 = nested_dict_to_df(rel_diff_dict_100y_not_caped).reset_index()
df2 = df2.rename(columns={'level_0':'Exposure','level_1':'Warming', "imp":"Full recovery"})
df2["Full recovery"] = df2["Full recovery"]*100


df3 = nested_dict_to_df(rel_diff_dict_100y_not_ordered).reset_index()
df3 = df3.rename(columns={'level_0':'Exposure','level_1':'Warming', "imp":"No common drivers"})
df3["No common drivers"] = df3["No common drivers"]*100


df = df.merge(df2)
df = df.merge(df3)
```
